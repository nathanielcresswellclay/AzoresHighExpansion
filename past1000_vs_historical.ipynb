I'd also like to include a montecarlo simulation where I find like probabiltiy that randomly selecting centuries from each model produces the same or more extreme events then artificially selected the last 100 years from each model. I have old script I wrote to do exactly this analysis. First I'll put the data into a form that is digestible by the script.

```python
bin_bcc   = bins['BCC-CSM1.1']
bin_miroc = bins['MIROC-ESM']
bin_ipsl  = bins['IPSL']
bin_mpi   = bins['MPI-ESM-P']
bin_ccsm  = bins['CCSM4']
```

...then I run the script

```python
#Assign parameters
iterations = 100000
percent = 10
window = 100

sample_industrial = True

###########################################################################################################################

#initialize array to hold extremely large events 
num_large = np.zeros([iterations])

#perform Monte Carlo sampling
for i in range(0,iterations):
           
    start = np.random.randint(50,1105)
    num_large[i] = num_large[i]+bin_bcc[start]
        
    start = np.random.randint(50,1105)
    num_large[i] = num_large[i]+bin_miroc[start]

    start = np.random.randint(50,1105)
    num_large[i] = num_large[i]+bin_ipsl[start]

    start = np.random.randint(50,1105)
    num_large[i] = num_large[i]+bin_mpi[start]

    start = np.random.randint(50,1105)
    num_large[i] = num_large[i]+bin_ccsm[start]
       
#sample post 1850 period from each ensemble member
ind_large = np.zeros([5])

ind_large[0] = bin_miroc[1105]
ind_large[1] = bin_bcc[1105]
ind_large[2] = bin_ipsl[1105]
ind_large[3] = bin_mpi[1105]
ind_large[4] = bin_ccsm[1105]
```

```python
%run AzoresHighProjectImports.py
%run /home/ncresswell/AzoresHighExpansion/MyFunctions/DataAnalyzing.py
%run /home/ncresswell/AzoresHighExpansion/MyFunctions/DataProcessing.py
%run /home/ncresswell/AzoresHighExpansion/MyFunctions/Plotting.py

from scipy.stats import ttest_ind
from scipy import stats

def get_aha(filename, lon_name='lon',lat_name='lat',var_name='psl',lat_range=[10,52],lon_range=[-60,10],threshold=1021.5,conversion_factor=1/100,mf_dataset = False,return_psl=False,
            detrend = True, trend_region = {'lon_min':-180,'lon_max':180,'lat_min':-90,'lat_max':90}):
    
    #load data
    if mf_dataset:
        psl = xr.open_mfdataset(filename,
                                decode_times=False,
                                decode_cf=False,
                                concat_dim='time')[var_name]
    else:
        psl = xr.open_dataset(filename)[var_name]
    #get seasonal average DJF, also center longitude
    psl_djf,centered_lon = fix_lon(seasonal_averages(psl.values,
                                                     season='DJF')*conversion_factor,
                                   psl[lon_name].values)
    
    if detrend:
        psl_djf = remove_trend_from_region(psl_djf,psl[lat_name],psl[lon_name],trend_region)
        
    #calculate AHA diagnostic
    ah = filter_lat_lon(data_unfiltered=filter_tresh(data_unfiltered=psl_djf,
                                                     below=True,
                                                     threshold=threshold),
                        lat=psl[lat_name].values,
                        lat_range=lat_range,
                        lon=centered_lon,
                        lon_range=lon_range)
    #find AHA
    aha = get_area_of_filtered_data(ah,lat=psl[lat_name].values,
                                    lon=centered_lon)
    if return_psl:
        return aha, psl_djf, centered_lon
    else:
        return aha  
    
def remove_trend_from_region(psl_djf,lat,lon,trend_region):
    
    dt_data = psl_djf[:,np.logical_and( lat>=trend_region['lat_min'], lat<=trend_region['lat_max']),:]
    dt_data = dt_data[:,:,np.logical_and( lon>=trend_region['lon_min'], lon<=trend_region['lon_max'])] 

    dt_time = np.arange(1850,1850+len(psl_djf))
    dt_timesteps = np.arange(0,dt_time.size)

    trends = np.empty(dt_data.shape[1:])

    #loop through space to get trends 
    for i in range(0,dt_data.shape[1]):
        for j in range(0,dt_data.shape[2]):    
            trends[i,j] = stats.linregress(dt_time,dt_data[:,i,j]).slope

    #create copy of array 
    psl_detrended = psl_djf.copy()

    #calculate geographic weights for later weighted avhc2ging ~~~~ again you must slice one dimension at a time
    weights = np.asarray([np.cos(np.deg2rad(lat))]*lon.shape[0]).transpose()[np.logical_and( lat>=trend_region['lat_min'], lat<=trend_region['lat_max']),:]
    weights = weights[:,np.logical_and( lon>=trend_region['lon_min'], lon<=trend_region['lon_max'])]

    dt_trend = np.average(trends, weights=weights)*dt_timesteps
    
    #loop through PSL and subtract trend_timeseries
    for la in np.where(np.logical_and( lat>trend_region['lat_min'], lat<trend_region['lat_max']))[0]:
        for lo in np.where(np.logical_and( lon>trend_region['lon_min'], lon<trend_region['lon_max']))[0]:

            psl_detrended[:,la,lo] = psl_djf[:,la,lo] - dt_trend

    return psl_detrended
    
def extreme_events(series1=None, series2=None, 
                   date_range1=np.arange(850,1850), date_range2=np.arange(1850,2006),
                   percent=10, window=101,
                   ax = None):
    
    concatenated=np.concatenate([series1, series2])
    
    highest,where_highest = get_extremes(series=concatenated,
                                         percent=percent,
                                         top=True);
    running_bin_big,_  = get_bin_timeseries(index=concatenated,
                                            percent=percent,
                                            window=window);
    if ax is None:
        fig,ax_ts = plt.subplots(figsize=(15,5))
    else: 
        ax_ts = ax
        fig=ax.figure

    ax_ts.plot(date_range1,series1,color='black',alpha=.3)
    ax_ts.plot(date_range2,series2,color='red',alpha=.3)
    ax_ts.scatter(where_highest+date_range1[0],highest,color='red',marker='^',alpha=.3)

    #plot ensemble mean extreme AHA running bin analysis
    ax_x=ax_ts.twinx()
    ax_x.plot(np.concatenate([date_range1,date_range2]),
              running_bin_big,color='red',linewidth=3,alpha=.5);

    #change labes sizes for ease of viewing
    ax_ts.set_xlim([date_range1[0],date_range2[-1]])
    ax_ts.set_xlabel('Year',fontsize=22)
    #larger dates 
    ax_ts.tick_params(labelsize=20)
    ax_ts.yaxis.offsetText.set_fontsize(15)
    # #larger yticks
    ax_x.tick_params(labelsize=20,colors='red')
            
    return fig, (ax_ts,ax_x)

def histogram_comparison(series1=None,series2=None,ax=None):
    
    if ax is None:
        fig,ax1 = plt.subplots(figsize=(10,5))
    else:
        fig=ax.figure
        ax1 = ax
        
    ax1.hist(series1,color='grey',alpha=.4,bins=20,edgecolor='black',rwidth=.5,align='mid');

    ax2 = ax1.twinx()
    ax2.hist(series2,color='lightcoral',alpha=.3,bins=20,edgecolor='red',rwidth=.5,align='mid');

    #change labes sizes for ease of viewing
    ax1.tick_params(labelsize=20)
    ax1.xaxis.offsetText.set_fontsize(15)
    ax2.tick_params(labelsize=20,colors='red')

    #ttest 
#     p = ttest_ind(series1,series2)[1]
#     ax1.text(.1,.5,'p='+str(np.around(p,3)), transform=ax1.transAxes,fontsize=20)
        
    return fig, (ax1,ax2)


def compare_extremes_hist(filename1=None,filename2=None,
                          mfdataset_file1=False,mfdataset_file2=False,
                          date_range1=np.arange(850,1851),date_range2=np.arange(1850,2006),
                          threshold=1021.5,
                          ax_series = None, ax_h = None):
    

    series1 = get_aha(filename1, mf_dataset=mfdataset_file1, threshold=threshold)
    series2 = get_aha(filename2, mf_dataset=mfdataset_file2, threshold=threshold)

    
    #make new plot fo extreme/timeseries figure of add plot to existing axes
    if ax_series is None:
        if date_range1[-1]==date_range2[0]:
            fig_extremes, axs_series = extreme_events(series1=series1[:-1],series2=series2,
                                                      date_range1=date_range1[:-1],date_range2=date_range2)
        else:
            fig_extremes, axs_series = extreme_events(series1=series1,series2=series2,
                                                     date_range1=date_range1,date_range2=date_range2)
    else:    
        if date_range1[-1]==date_range2[0]:
            fig_extremes, axs_series = extreme_events(series1=series1[:-1],series2=series2,
                                                      date_range1=date_range1[:-1],date_range2=date_range2,
                                                      ax=ax_series)
        else:
            fig_extremes, axs_series = extreme_events(series1=series1,series2=series2,
                                                      date_range1=date_range1,date_range2=date_range2,
                                                      ax=ax_series)
            
    #make new plot for histogram figure or draw on existing plot
    if ax_h is None:
        fig_hist,axs_h = histogram_comparison(series1=series1,series2=series2)
    else:
        fig_hist,axs_h = histogram_comparison(series1=series1,series2=series2,ax=ax_h)
    
    return (fig_extremes,axs_series) , (fig_hist, axs_h)

def add_row_label(ax,text):
    ax.text(-0.1, 0.55, text, va='bottom', ha='center',
                      rotation='vertical', rotation_mode='anchor',
                      transform=ax.transAxes,fontsize=30)

"""
This function will plot frequency of extreme events as reported by ERA-20C pressure data over a defined window 

USAGE: to use pass an index that is defined through time, and specify what the threshold is for extreme event using the
      'percent' perameter. 

"""

def running_bin_extremes(index=None, percent=10, window=101, time=np.arange(1850,2006), label_large = '',label_small = '',title = '',legend_loc=2, index_has_nan=False,ylim=None):

    running_bin_big   = np.zeros(index.shape)
    running_bin_small = np.zeros(index_shape)
    
    running_bin_big, running_bin_small = get_bin_timeseries(index=index, percent=percent, window=window)
        
    fig, ax = plt.subplots(figsize=(20,10))

    ax.plot(time,running_bin_big,color='red',linewidth=3,alpha=2,label=label_large);
    ax.plot(time,running_bin_small,color='blue',linewidth=3,alpha=10,label=label_small);
   
    #Alter Ticks and labels
    ax.set_xticklabels(ax.get_xticks().astype(int),fontsize=20)
    ax.set_yticklabels(ax.get_yticks().astype(int),fontsize=20)
    ax.set_ylabel('Number of Extreme Events',fontsize=25)
    
    ax.legend(fontsize=30,loc=legend_loc);
    ax.set_title(title,fontsize=35);
    
    return fig,ax

def get_running_bin_past1000_hist(past1000_filename=None,                 historical_filename=None,
                                  past1000_mfdata=False,                  historical_mfdata=False,
                                  past1000_daterange=np.arange(850,1851), historical_daterange=np.arange(1850,2006),
                                  threshold=1021.5,percent=10,window=101):
    
    #get_aha from each of your files
    past1000_aha   = get_aha(past1000_filename,   mf_dataset=past1000_mfdata,   threshold=threshold)
    historical_aha = get_aha(historical_filename, mf_dataset=historical_mfdata, threshold=threshold)
    
    #concatenate time series -- check for overlap
    if past1000_daterange[-1]==historical_daterange[0]:
        concatenated=np.concatenate([past1000_aha, historical_aha[1:]])
    else:
        concatenated=np.concatenate([past1000_aha, historical_aha])
  
    highest,where_highest = get_extremes(series=concatenated,percent=percent,top=True);
    running_bin_big,_     = get_bin_timeseries(index=concatenated,percent=percent,window=window);
    
    return running_bin_big 

def get_running_bin_past1000_hist_BULK(dicts):
    
    #initialize empty dictionary
    running_bins = {}
    
    #loop through past dictionaries 
    for i in range(0,len(dicts)):
        
        print('working on '+dicts[i]['model']+'...')
        
        temp =  get_running_bin_past1000_hist(past1000_filename    =dicts[i]['past1000_filename'],
                                              historical_filename  =dicts[i]['historical_filename'],
                                              past1000_mfdata      =dicts[i]['past1000_mfdata'],
                                              historical_mfdata    =dicts[i]['historical_mfdata'],
                                              past1000_daterange   =dicts[i]['past1000_daterange'],
                                              historical_daterange =dicts[i]['historical_daterange'],
                                              threshold            =dicts[i]['threshold']) 
        #save the calculated series with the model name
        running_bins.update({dicts[i]['model']:temp})
        print('...'+'DONE!')
    
    return running_bins

"""
This function offers the same functionality of the running_bin_extremes_era data but is altered to account for the CESM_LME
meaning it assumed the first dimension of the data is ensemble member number. 

usage: pass index defined along ensemble member and time. observe the default parameters used in the calculation and
       rendering of the plot and change if desired 
"""

def running_bin_extremes_lme(index=None, percent=20, window=11, label_large = '',label_small = '',title = '',legend_loc=2, index_has_nan=False,ylim=None):

    #initialize containers for running bin diagnostic 
    running_bin_big   = np.zeros(index.shape)
    running_bin_small = np.zeros(index.shape)
    
    #iterate through ensemble members
    for e_mem in range(0,13):

        running_bin_big[e_mem,:],running_bin_small[e_mem,:] = get_bin_timeseries(index=index[e_mem,:], 
                                                                                 percent=percent, window=window)

    fig, ax = plt.subplots(figsize=(20,10))
        
    ax.plot(time,running_bin_big.mean(axis=0),color='red',linewidth=3,alpha=2,label=label_large);
    ax.plot(time,running_bin_small.mean(axis=0),color='blue',linewidth=3,alpha=10,label=label_small);
   
    #Alter Ticks and labels
    ax.set_xticklabels(ax.get_xticks().astype(int),fontsize=20)
    ax.set_yticklabels(ax.get_yticks().astype(int),fontsize=20)
    ax.set_ylabel('Number of Extreme Events',fontsize=25)
    
    ax.legend(fontsize=30,loc=legend_loc);
    ax.set_title(title,fontsize=35);
    
    return fig,ax

def get_bin_timeseries(index=None, percent=20, window=11):
    
    running_bin_big   = np.zeros(index.shape)
    running_bin_small = np.zeros(index.shape)
    
    #get list of extreme years
    _,where_largest  = get_extremes(series=index,percent=percent,top=True)
    _,where_smallest = get_extremes(series=index,percent=percent,bot=True)
    
    running_bin_big,_  = extremes_per_window(date_range=np.arange(0,index.size), \
                                                      dates_of_extremes=where_largest, \
                                                      window=window);
    running_bin_small,_= extremes_per_window(date_range=np.arange(0,index.size), \
                                                      dates_of_extremes=where_smallest, \
                                                      window=window);
    
    return running_bin_big, running_bin_small

def add_subplot_index(ax,index='a',y=.87,x=.025):
    ax.text(x=x,y=y,
            s=index,
            fontsize=30,
            transform=ax.transAxes,fontweight='bold')
```
